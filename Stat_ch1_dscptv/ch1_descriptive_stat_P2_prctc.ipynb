# Exercise: Skewness

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from collections import Counter


data = [212, 869, 220, 654, 511, 624, 420, 121, 428, 865, 799, 405, 230, 670, 870, 366, 99, 55, 489, 312, 493, 163, 221, 84, 144, 48, 375, 86, 168, 100]

# Calculate mean, median, mode
mean_val = np.mean(data)
median_val = np.median(data)
mode_val = Counter(data).most_common(1)[0][0]


plt.figure(figsize=(9, 5))
ax = sns.histplot(data, bins=5, kde=True, color="#6495ed", edgecolor='white', alpha=0.6, shrink=0.9)

# Increase KDE line thickness
for line in ax.lines:
    line.set_linewidth(3)  # set thickness here


# Add vertical lines
# Add vertical lines
plt.axvline(mean_val, color='#e52b52', linestyle='-', linewidth=8, label=f"Mean: {mean_val:.1f}")
plt.axvline(median_val, color="#007fff", linestyle='-', linewidth=5, label=f"Median: {median_val:.1f}")
plt.axvline(mode_val, color="#9a9a9a", linestyle='-', linewidth=2, label=f"Mode: {mode_val}")


plt.title("Dataset 1 Distribution", fontsize=16, color="#54626f")
plt.xlabel("Value", fontsize=16, color="#8cc061")
plt.ylabel("Frequency", fontsize=16, color="#87a96b")

plt.xticks(fontsize=16, color="#54626f")
plt.yticks(fontsize=16, color="#54626f")
plt.grid(axis="y", linestyle="--", linewidth=0.5, alpha=0.8, color="#72a0c1")

for spine in plt.gca().spines.values():
    spine.set_visible(False)

plt.legend(loc="upper right", fontsize=16)

plt.tight_layout()
plt.show()
```

# Exercise: Co-varience, Correlation coefficient

```python
import matplotlib.pyplot as plt

# Data
writing = [344, 383, 611, 713, 536]
reading = [378, 349, 503, 719, 503]



# Font name
fntNm = "Calibri"

# Create scatter plot
plt.figure(figsize=(4, 4))
plt.scatter(writing, reading,
            color="#89cff0", edgecolors="#318ce7",
            s=60, alpha=0.85)

# Mean points
x_mean = 517.4
y_mean = 490.4
plt.scatter(x_mean, y_mean, color="red", s=70, marker="X", label="Mean Point (517.4, 490.4)")

# Axis limits (with small margin)
plt.xlim(200, 800)
plt.ylim(200, 800)

# Labels and title
plt.title("Writing vs. Reading", fontsize=15, color="#54626f", pad=15, fontname=fntNm)
plt.xlabel("Writing score", fontsize=12, color="#87a96b", fontname=fntNm)
plt.ylabel("Reading score", fontsize=12, color="#87a96b", fontname=fntNm)

# Tick style
plt.xticks(color="#54626f", fontsize=10)
plt.yticks(color="#54626f", fontsize=10)

# Grid and spines
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.6, color="#72a0c1")
for spine in plt.gca().spines.values():
    spine.set_visible(False)

# Add legend
plt.legend(fontsize=9)

# Place the legend outside the plot
plt.legend(bbox_to_anchor=(0.1, -0.2), loc='upper left', borderaxespad=0.)

plt.tight_layout()
plt.show()
```

# Exercise : "Practiacal Example- Cal Real State"

```python
# clean the old CSV and save to new

import pandas as pd

# Load raw file (no header so we can rebuild it)
df_raw = pd.read_csv("./ch1_data_CAL_real_state.csv", header=None)

# Step 1: Use first row as column names
new_columns = df_raw.iloc[0]

# Step 2: Assign proper column names
df = df_raw[1:].copy()
df.columns = new_columns

# Step 3: Reset index
df.reset_index(drop=True, inplace=True)

# Step 4: Remove completely empty columns
df = df.dropna(axis=1, how='all')

# Step 5: Clean column names (strip spaces)
df.columns = df.columns.str.strip()

# # Step 6: Optional — rename duplicate / unclear columns
# df.columns = pd.io.parsers.ParserBase({'names': df.columns})._maybe_dedup_names(df.columns)

# Step 7: Clean price column (remove $ and commas → float)
if 'Price' in df.columns:
    df['Price'] = (
        df['Price']
        .replace('[\$,]', '', regex=True)
        .astype(float)
    )

# Step 8: Convert numeric columns where possible
df = df.apply(pd.to_numeric, errors='ignore')
```

```python
# Display the DataFrame
print(df)
```

```python
# Step 9: Save clean CSV
df.to_csv("CalRealState_data_cleaned.csv", index=False)

print("Cleaned file saved as cleaned_data.csv")
```

```python
import matplotlib.pyplot as plt

# Data
gender = ["Male", "Female", "Firms"]
sales = [108, 70, 17]
colors = ["#3188e6ff", "#3cd198ff", "#f26a2bff"]  # Muted, modern colors

# Create the bar chart
plt.figure(figsize=(8, 6))
bars = plt.bar(gender, sales, color=colors, edgecolor='none', width=0.7)

# Add labels and title
plt.xlabel("Gender", fontsize=12, color="#87a96b")
plt.ylabel("Frequency", fontsize=12, color="#87a96b")
plt.title("Gender Distribution", fontsize=16, color="#54626f")

# Display values on top of bars
for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 4, 
             str(bar.get_height()), ha="center", va="bottom", fontsize=12, color='#6495ed')

# show grids
plt.grid(axis='y', linestyle="--", alpha=0.6)  # Enables grid with dashed lines and transparency
plt.ylim(0, 120)  # Adjust the Y-axis


# Remove top and right spines
ax = plt.gca()  # get current Axes
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

# Show the chart
plt.show()
```
