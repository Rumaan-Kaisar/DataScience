```python
# clean the old CSV and save to new

import pandas as pd

# Load raw file (no header so we can rebuild it)
df_raw = pd.read_csv("./ch1_data_CAL_real_state.csv", header=None)

# Step 1: Use first row as column names
new_columns = df_raw.iloc[0]

# Step 2: Assign proper column names
df = df_raw[1:].copy()
df.columns = new_columns

# Step 3: Reset index
df.reset_index(drop=True, inplace=True)

# Step 4: Remove completely empty columns
df = df.dropna(axis=1, how='all')

# Step 5: Clean column names (strip spaces)
df.columns = df.columns.str.strip()

# # Step 6: Optional — rename duplicate / unclear columns
# df.columns = pd.io.parsers.ParserBase({'names': df.columns})._maybe_dedup_names(df.columns)

# Step 7: Clean price column (remove $ and commas → float)
if 'Price' in df.columns:
    df['Price'] = (
        df['Price']
        .replace('[\$,]', '', regex=True)
        .astype(float)
    )

# Step 8: Convert numeric columns where possible
df = df.apply(pd.to_numeric, errors='ignore')
```

```python
# Step 9: Save clean CSV
df.to_csv("CalRealState_data_cleaned.csv", index=False)

print("Cleaned file saved as cleaned_data.csv")
```

```python
import pandas as pd

# -----------------------------
# State frequency analysis
# -----------------------------

# Get all state values (excluding NaN)
state_values = customer["State"].dropna().unique().tolist()

# Count total observations
total_all = len(customer)

# Count only US customers
total_us = customer["State"].dropna().count()

# Dictionary for frequencies
state_count = {}

for state in state_values:
    state_count[state] = customer[customer["State"] == state].shape[0]

# Add abroad / missing values
none_abroad = customer["State"].isna().sum()
state_count["None (abroad)"] = none_abroad

# Create dataframe
freq_dist = pd.DataFrame({
    "State": state_count.keys(),
    "Frequency": state_count.values()
})

# Sort by frequency descending
freq_dist = freq_dist.sort_values(by="Frequency", ascending=False).reset_index(drop=True)

# Relative frequency (all customers)
freq_dist["Relative freq."] = (
    (freq_dist["Frequency"] / total_all) * 100
).round().astype(int).astype(str) + "%"

# Relative frequency (US only)
freq_dist["Relative US only"] = ""

mask = freq_dist["State"] != "None (abroad)"

freq_dist.loc[mask, "Relative US only"] = (
    (freq_dist.loc[mask, "Frequency"] / total_us) * 100
).round().astype(int).astype(str) + "%"

# Cumulative frequency (all)
cum_all = (
    freq_dist["Frequency"].cumsum() / total_all * 100
).round().astype(int).astype(str) + "%"

freq_dist["Cumulative freq."] = cum_all

# Cumulative frequency (US only)
us_cumsum = (
    freq_dist.loc[mask, "Frequency"].cumsum() / total_us * 100
).round().astype(int).astype(str) + "%"

freq_dist["Cumulative US only"] = ""
freq_dist.loc[mask, "Cumulative US only"] = us_cumsum.values

# Add total row
total_row = pd.DataFrame([{
    "State": f"Total US = {total_us}",
    "Frequency": f"Total = {total_all}",
    "Relative freq.": "100%",
    "Relative US only": "",
    "Cumulative freq.": "",
    "Cumulative US only": ""
}])

freq_dist = pd.concat([freq_dist, total_row], ignore_index=True)

# Display table
print(freq_dist)
```

```python
# Print table using TAB spacing
print("State\tFrequency\tRelative freq.\tRelative US only\tCumulative freq.\tCumulative US only")

for i in range(len(freq_dist)):
    row = freq_dist.iloc[i]
    
    print(
        f"{row['State']}\t"
        f"{row['Frequency']}\t"
        f"{row['Relative freq.']}\t"
        f"{row['Relative US only']}\t"
        f"{row['Cumulative freq.']}\t"
        f"{row['Cumulative US only']}"
    )
```

```python
import matplotlib.pyplot as plt
import numpy as np

# -----------------------------
# Prepare data from frequency table
# -----------------------------

# Remove total row
pareto_df = freq_dist[
    ~freq_dist["State"].astype(str).str.contains("Total")
].copy()

# Remove "None (abroad)" if desired
# pareto_df = pareto_df[pareto_df["State"] != "None (abroad)"]

# States and frequencies
states = pareto_df["State"].tolist()
freqs = pareto_df["Frequency"].astype(int).tolist()

# Sort descending
sorted_indices = np.argsort(freqs)[::-1]

sorted_states = [states[i] for i in sorted_indices]
sorted_freqs = [freqs[i] for i in sorted_indices]

# Spectrum colors (high → low)
# colors = plt.cm.Spectral(np.linspace(0.1, 0.9, len(sorted_states)))
colors = plt.cm.Blues(np.linspace(0.4, 0.9, len(sorted_states)))

# Cumulative percentage
cumulative = np.cumsum(sorted_freqs)
cumulative_percent = cumulative / sum(sorted_freqs) * 100

# -----------------------------
# Plot
# -----------------------------

fig, ax1 = plt.subplots(figsize=(10, 5))

# Bar chart
bars = ax1.bar(
    sorted_states,
    sorted_freqs,
    color=colors,
    edgecolor="white"
)

# Labels and title
ax1.set_ylabel("Frequency", fontsize=11, color="#54626f")
ax1.set_title(
    "Pareto Diagram: Customer States",
    fontsize=14,
    color="#3e9df6",
    pad=15
)

# Ticks
ax1.tick_params(axis='x', labelrotation=25, labelsize=10, colors="#117de2")
ax1.tick_params(axis='y', labelsize=10, colors="#54626f")

# Grid
ax1.grid(axis="y", linestyle="--", linewidth=0.5, alpha=0.7)

# Remove spines
for spine in ax1.spines.values():
    spine.set_visible(False)

# Frequency labels on bars
for bar in bars:
    height = bar.get_height()

    ax1.text(
        bar.get_x() + bar.get_width()/2,
        height + 10,
        str(int(height)),
        ha="center",
        va="bottom",
        fontsize=10,
        color="#ff2052"
    )

# -----------------------------
# Cumulative line
# -----------------------------

ax2 = ax1.twinx()

ax2.plot(
    sorted_states,
    cumulative_percent,
    color="#ff2052",
    marker="o",
    linewidth=2
)

ax2.set_ylabel(
    "Cumulative %",
    fontsize=11,
    color="#ff2052"
)

ax2.set_ylim(0, 110)

ax2.tick_params(
    axis='y',
    colors="#ff2052",
    labelsize=10
)

# 80% reference line
ax2.axhline(
    80,
    color="gray",
    linestyle="--",
    linewidth=1
)

# Cumulative % labels
for i, val in enumerate(cumulative_percent):

    ax2.text(
        i,
        val + 3,
        f"{val:.1f}%",
        ha="center",
        va="center",
        fontsize=9,
        color="#444444"
    )

# Remove spines
for spine in ax2.spines.values():
    spine.set_visible(False)

plt.tight_layout()
plt.show()
```
