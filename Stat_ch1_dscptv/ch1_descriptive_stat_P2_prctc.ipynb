# Exercise: Skewness

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from collections import Counter


data = [212, 869, 220, 654, 511, 624, 420, 121, 428, 865, 799, 405, 230, 670, 870, 366, 99, 55, 489, 312, 493, 163, 221, 84, 144, 48, 375, 86, 168, 100]

# Calculate mean, median, mode
mean_val = np.mean(data)
median_val = np.median(data)
mode_val = Counter(data).most_common(1)[0][0]


plt.figure(figsize=(9, 5))
ax = sns.histplot(data, bins=5, kde=True, color="#6495ed", edgecolor='white', alpha=0.6, shrink=0.9)

# Increase KDE line thickness
for line in ax.lines:
    line.set_linewidth(3)  # set thickness here


# Add vertical lines
# Add vertical lines
plt.axvline(mean_val, color='#e52b52', linestyle='-', linewidth=8, label=f"Mean: {mean_val:.1f}")
plt.axvline(median_val, color="#007fff", linestyle='-', linewidth=5, label=f"Median: {median_val:.1f}")
plt.axvline(mode_val, color="#9a9a9a", linestyle='-', linewidth=2, label=f"Mode: {mode_val}")


plt.title("Dataset 1 Distribution", fontsize=16, color="#54626f")
plt.xlabel("Value", fontsize=16, color="#8cc061")
plt.ylabel("Frequency", fontsize=16, color="#87a96b")

plt.xticks(fontsize=16, color="#54626f")
plt.yticks(fontsize=16, color="#54626f")
plt.grid(axis="y", linestyle="--", linewidth=0.5, alpha=0.8, color="#72a0c1")

for spine in plt.gca().spines.values():
    spine.set_visible(False)

plt.legend(loc="upper right", fontsize=16)

plt.tight_layout()
plt.show()
```

# Exercise: Co-varience, Correlation coefficient

```python
import matplotlib.pyplot as plt

# Data
writing = [344, 383, 611, 713, 536]
reading = [378, 349, 503, 719, 503]



# Font name
fntNm = "Calibri"

# Create scatter plot
plt.figure(figsize=(4, 4))
plt.scatter(writing, reading,
            color="#89cff0", edgecolors="#318ce7",
            s=60, alpha=0.85)

# Mean points
x_mean = 517.4
y_mean = 490.4
plt.scatter(x_mean, y_mean, color="red", s=70, marker="X", label="Mean Point (517.4, 490.4)")

# Axis limits (with small margin)
plt.xlim(200, 800)
plt.ylim(200, 800)

# Labels and title
plt.title("Writing vs. Reading", fontsize=15, color="#54626f", pad=15, fontname=fntNm)
plt.xlabel("Writing score", fontsize=12, color="#87a96b", fontname=fntNm)
plt.ylabel("Reading score", fontsize=12, color="#87a96b", fontname=fntNm)

# Tick style
plt.xticks(color="#54626f", fontsize=10)
plt.yticks(color="#54626f", fontsize=10)

# Grid and spines
plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.6, color="#72a0c1")
for spine in plt.gca().spines.values():
    spine.set_visible(False)

# Add legend
plt.legend(fontsize=9)

# Place the legend outside the plot
plt.legend(bbox_to_anchor=(0.1, -0.2), loc='upper left', borderaxespad=0.)

plt.tight_layout()
plt.show()
```
