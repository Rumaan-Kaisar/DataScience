````python
import matplotlib.pyplot as plt

# Data
cups = [0, 1, 2, 3, 4, 5]
people = [5, 20, 40, 25, 8, 2]

# Create histogram-like bar chart
plt.figure(figsize=(8, 5))
bars = plt.bar(cups, people, color="#00c4b0", edgecolor="white", width=0.8)

# Titles and labels
plt.title("Distribution of Coffee Consumption", fontsize=15, color="#54626f", pad=15)
plt.xlabel("Cups per Day", fontsize=12, color="#87a96b")
plt.ylabel("Number of People", fontsize=12, color="#87a96b")

# Ticks and grid
plt.xticks(cups, fontsize=10, color="#54626f")
plt.yticks(fontsize=10, color="#54626f")
plt.grid(axis="y", linestyle="--", linewidth=0.5, alpha=0.6, color="#72a0c1")

# Minimalist styling
for spine in plt.gca().spines.values():
    spine.set_visible(False)

plt.tight_layout()
plt.show()
````

## ðŸ”‘ Key Concepts in Data Distributions (with Python Examples)

---

### **1. Symmetric (Normal) Distribution**

* **Shape**: Bell-shaped and balanced around the center.
* **Example**: Heights of adults.
* **Why it matters**:

  * Mean â‰ˆ Median â‰ˆ Mode.
  * Predictable spread: \~68% data within 1 standard deviation.

#### ðŸ”¹ Python Code:

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = np.random.normal(loc=170, scale=10, size=1000)  # Simulated adult heights

plt.figure(figsize=(8, 5))
sns.histplot(data, bins=30, kde=True, color="#4c9f70")
plt.title("Symmetric (Normal) Distribution - Adult Heights", fontsize=14)
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

---

### **2. Skewed Distributions**

#### a. **Right-Skewed (Positively Skewed)**

* **Tail** extends to the right.
* **Example**: Income levels (few people are ultra-rich).
* **Feature**: Mean > Median.

#### ðŸ”¹ Python Code:

```python
data = np.random.exponential(scale=20000, size=1000) + 30000  # Simulated income

plt.figure(figsize=(8, 5))
sns.histplot(data, bins=30, kde=True, color="#ff8c42")
plt.title("Right-Skewed Distribution - Income Levels", fontsize=14)
plt.xlabel("Income ($)")
plt.ylabel("Frequency")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

#### b. **Left-Skewed (Negatively Skewed)**

* **Tail** extends to the left.
* **Example**: Retirement ages (most retire around 60â€“65, few much earlier).
* **Feature**: Mean < Median.

#### ðŸ”¹ Python Code:

```python
data = 70 - np.random.exponential(scale=5, size=1000)  # Simulated retirement ages
data = np.clip(data, 30, 70)

plt.figure(figsize=(8, 5))
sns.histplot(data, bins=30, kde=True, color="#6495ed")
plt.title("Left-Skewed Distribution - Retirement Age", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

---

### **3. Spread (Variance / Standard Deviation)**

* **Measures**: How spread out the values are.
* **Low SD Example**: Scores like 75, 76, 77, 78 (clustered).
* **High SD Example**: Scores like 30, 60, 90, 100 (scattered).

#### ðŸ”¹ Python Code:

```python
low_spread = [75, 76, 77, 78, 76, 77, 75, 78]
high_spread = [30, 60, 90, 100, 45, 95, 35, 80]

fig, axs = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

sns.histplot(low_spread, bins=5, kde=True, ax=axs[0], color="#88d8b0")
axs[0].set_title("Low Spread")
axs[0].set_xlabel("Score")
axs[0].grid(True, linestyle="--", alpha=0.5)

sns.histplot(high_spread, bins=5, kde=True, ax=axs[1], color="#f67280")
axs[1].set_title("High Spread")
axs[1].set_xlabel("Score")
axs[1].grid(True, linestyle="--", alpha=0.5)

plt.suptitle("Spread Comparison: Low vs High Standard Deviation", fontsize=14)
plt.tight_layout()
plt.show()
```

## ðŸ’¡ Why Distributions Matter in Statistics

| Concept              | What It Tells You                  |
| -------------------- | ---------------------------------- |
| **Mean/Median/Mode** | Center of the data                 |
| **Skewness**         | Lopsidedness in the data shape     |
| **Variance/SD**      | Width or spread of the values      |
| **Correlation**      | Relationship between two variables |

---

### ðŸ§  Real-World Analogies

* **Heights**: Normal distribution â€” most people are around average.
* **Wealth**: Right-skewed â€” few rich, many with average income.
* **Spread**: Clustered group (low SD) vs scattered (high SD).

## âœ… Summary

* **Distribution** = The pattern of your data values.
* **Normal** = Symmetric; mean â‰ˆ median â‰ˆ mode.
* **Skewed** = Lopsided; mean pulled by outliers.
* **Spread** = How far values deviate from the center.
