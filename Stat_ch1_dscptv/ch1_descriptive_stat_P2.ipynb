The issue is that the KDE `y_vals` are density estimates that integrate to 1, while the histogram is plotted using counts (frequencies). So the KDE curve scale is not matching the histogram's y-axis (frequency scale).

To fix this, you need to scale the KDE values so that its area matches the histogram's total area. The common approach is:

**Scale KDE by:**
`kde_scaled = kde(x_vals) * total_number_of_points * bin_width`

Here's how to fix your code:

1. Find the histogram bin width (you used `bins=range(4, 74)`, so bin width is 1).
2. Multiply `y_vals` by `len(prices)` (total data points) and `bin_width`.

Modify your KDE scaling as:

```python
bin_width = 1
y_vals = kde(x_vals) * len(prices) * bin_width
```

Full relevant part updated:

```python
# KDE calculation and scaling
kde = gaussian_kde(prices)
x_vals = np.linspace(4, 74, 500)
bin_width = 1
y_vals = kde(x_vals) * len(prices) * bin_width
```

This will scale the KDE to match the histogram's frequency scale and align nicely on the plot.

````python
from brokenaxes import brokenaxes
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import gaussian_kde
from collections import Counter

# Data
prices = [4, 4, 5, 5, 6, 6, 6, 7, 7, 7, 8, 9, 9, 10, 72]

# KDE calculation
kde = gaussian_kde(prices)
x_vals = np.linspace(4, 74, 500)
# y_vals = kde(x_vals)

bin_width = 1
y_vals = kde(x_vals) * len(prices) * bin_width


# Create figure
fig = plt.figure(figsize=(10, 5))
bax = brokenaxes(xlims=((4, 35), (70, 74)), hspace=0.05)

# Histogram
bax.hist(prices, bins=range(4, 74), color="#4c9f70", edgecolor="white", alpha=0.6)

# KDE plot
bax.plot(x_vals, y_vals, color="#6495ed", linewidth=2, label="KDE")

# Mean, Median, Mode
mean = np.mean(prices)
median = np.median(prices)
mode = max(Counter(prices), key=Counter(prices).get)

bax.axvline(mean, color="red", linestyle="--", label=f"Mean: {mean:.1f}")
bax.axvline(median, color="green", linestyle="-", label=f"Median: {median}")
bax.axvline(mode, color="purple", linestyle=":", label=f"Mode: {mode}")

# Labels and styling
bax.set_title("Histogram of NYC Pizza Prices", fontsize=14)
bax.set_xlabel("Price ($)", fontsize=12, labelpad=20)
bax.set_ylabel("Frequency", fontsize=12, labelpad=30)
bax.legend(loc="upper right")

# To enable grid lines on a brokenaxes plot (bax), you cannot use plt.grid() as it only affects the default plt.gca() axes — not the custom sub-axes inside brokenaxes.
# Instead, you must loop through the individual axes inside the brokenaxes object. Here’s how to do it:
# Add this after your plotting code (but before plt.show()):

# Set grid for each subplot in brokenaxes
for ax in bax.axs:
    ax.grid(axis="y", linestyle="--", linewidth=0.5, alpha=0.6, color="#72a0c1")

    # Optional: Hide top/right spines for clean look
    for spine in ax.spines.values():
        spine.set_visible(False)


# plt.tight_layout()
plt.show()
````

To increase the KDE line thickness in your seaborn `histplot`, you can pass the `kde_kws` argument with a dictionary containing `linewidth` or `lw`. For example:

```python
sns.histplot(data, bins=7, kde=True, color="#6495ed", edgecolor='black', kde_kws={'linewidth': 3})
```
