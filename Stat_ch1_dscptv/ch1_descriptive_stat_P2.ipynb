## âœ… 1. Symmetric (Normal) Distribution

### ðŸ”¹ Dataset: Adult Heights (cm)

```python
import numpy as np

# Seed for reproducibility
np.random.seed(1)

# 1. Symmetric (Normal-like) dataset: Adult Heights
symmetric_data = np.random.normal(loc=170, scale=3, size=200).round(1)

# 2. Right-Skewed dataset: Incomes (log-normal)
right_skewed_data = np.random.lognormal(mean=3.5, sigma=0.5, size=200).round(1)

# 3. Left-Skewed dataset: Exam scores (simulate reverse right-skew)
left_skewed_data = (100 - np.random.lognormal(mean=3.5, sigma=0.5, size=200)).clip(min=0).round(1)

# 4. Low spread (clustered scores)
low_spread_data = np.random.normal(loc=75, scale=2, size=200).round(1)

# 5. High spread (scattered scores)
high_spread_data = np.random.normal(loc=75, scale=15, size=200).round(1)

# Optional: view a few samples
print("Symmetric (Heights):", symmetric_data[:5])
print("Right-Skewed (Incomes):", right_skewed_data[:5])
print("Left-Skewed (Exam Scores):", left_skewed_data[:5])
print("Low Spread:", low_spread_data[:5])
print("High Spread:", high_spread_data[:5])
```

```python
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
heights = [168, 170, 171, 169, 173, 172, 168, 170, 171, 170,
           172, 169, 171, 170, 173, 172, 170, 169, 171, 172]

# Calculate statistics
mean_val = np.mean(heights)
median_val = np.median(heights)
mode_val = Counter(heights).most_common(1)[0][0]

# Plot
plt.figure(figsize=(8, 5))
sns.histplot(heights, bins=5, kde=True, color="#4c909f", shrink=0.9, edgecolor='white')
plt.title("Symmetric Distribution: Adult Heights", fontsize=14)
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.xlim(167, 174)

# Add mean, median, mode lines
plt.axvline(mean_val, color="#1f77b4", linestyle="-", linewidth=4, label=f"Mean: {mean_val:.1f}")
plt.axvline(median_val, color="#2ca02c", linestyle="-", linewidth=4, label=f"Median: {median_val}")
plt.axvline(mode_val, color="#d62728", linestyle="-", linewidth=4, label=f"Mode: {mode_val}")

# Ticks and grid
plt.xticks(fontsize=10, color="#54626f")
plt.yticks(fontsize=10, color="#54626f")
plt.grid(axis="y", linestyle="--", linewidth=0.5, alpha=0.6, color="#72a0c1")

# Minimalist styling
for spine in plt.gca().spines.values():
    spine.set_visible(False)

plt.legend(loc="upper left", fontsize=10)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

# Sample right-skewed income data ($1000s)
income = [25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 80, 90, 100, 130, 150, 170, 200, 220, 250, 300]

# Calculate statistics
mean_val = np.mean(income)
median_val = np.median(income)
mode_val = Counter(income).most_common(1)[0][0]

# Plot
plt.figure(figsize=(8, 5))
sns.histplot(income, bins=5, kde=True, color="#ff8c42", edgecolor='white', shrink=0.9)
plt.title("Right-Skewed Distribution: Incomes", fontsize=14)
plt.xlabel("Income ($1000s)")
plt.ylabel("Frequency")

# Add vertical lines for mean, median, and mode
plt.axvline(mean_val, color="#1f77b4", linestyle="-", linewidth=4, label=f"Mean: {mean_val:.1f}")
plt.axvline(median_val, color="#2ca02c", linestyle="-", linewidth=4, label=f"Median: {median_val}")
plt.axvline(mode_val, color="#d62728", linestyle="-", linewidth=4, label=f"Mode: {mode_val}")

# Ticks and grid
plt.xticks(fontsize=10, color="#54626f")
plt.yticks(fontsize=10, color="#54626f")
plt.grid(axis="y", linestyle="--", linewidth=0.5, alpha=0.6, color="#72a0c1")

# Minimalist styling
for spine in plt.gca().spines.values():
    spine.set_visible(False)

plt.legend(loc="upper right", fontsize=10)
plt.tight_layout()
plt.show()
```

## âœ… 4. Spread (Low vs High Standard Deviation)

### ðŸ”¹ Dataset A: Low Spread (Clustered Scores)



```python
scores_low_sd = [75, 74, 76, 73, 77, 76, 75, 74, 75, 76]
```



### ðŸ”¹ Dataset B: High Spread (Scattered Scores)



```python
scores_high_sd = [60, 85, 90, 45, 100, 55, 65, 80, 70, 40]
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

# Subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

# Low Spread
sns.histplot(scores_low_sd, bins=4, kde=True, ax=axs[0], color="#88d8b0")
axs[0].set_title("Low Spread", fontsize=13)
axs[0].tick_params(axis='both', labelsize=10, colors="#54626f")
axs[0].grid(axis="y", linestyle="--", linewidth=0.5, alpha=0.6, color="#72a0c1")

# High Spread
sns.histplot(scores_high_sd, bins=4, kde=True, ax=axs[1], color="#f67280")
axs[1].set_title("High Spread", fontsize=13)
axs[1].tick_params(axis='both', labelsize=10, colors="#54626f")
axs[1].grid(axis="y", linestyle="--", linewidth=0.5, alpha=0.6, color="#72a0c1")

# Minimalist styling
for ax in axs:
    for spine in ax.spines.values():
        spine.set_visible(False)

plt.suptitle("Spread Comparison: Standard Deviation", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from collections import Counter

# Helper function to plot lines and labels
def add_stats_lines(ax, data):
    mean_val = np.mean(data)
    median_val = np.median(data)
    mode_val = Counter(data).most_common(1)[0][0]

    ymax = ax.get_ylim()[1]

    ax.axvline(mean_val, color='red', linestyle='-', linewidth=3.5)
    ax.axvline(median_val, color='green', linestyle='-', linewidth=3.5)
    ax.axvline(mode_val, color='purple', linestyle='-', linewidth=3.5)

    # ax.text(mean_val, ymax * 0.85, f"Mean\n{mean_val:.1f}", color='red', ha='center', fontsize=9)
    # ax.text(median_val, ymax * 0.65, f"Median\n{median_val:.1f}", color='green', ha='center', fontsize=9)
    # ax.text(mode_val, ymax * 0.45, f"Mode\n{mode_val}", color='purple', ha='center', fontsize=9)

# Subplots
fig, axs = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

# Low Spread
sns.histplot(scores_low_sd, bins=4, kde=True, ax=axs[0], color="#88d8b0", element="bars", shrink=0.93, edgecolor='white')
axs[0].set_title("Low Spread", fontsize=13)
axs[0].tick_params(axis='both', labelsize=10, colors="#54626f")
axs[0].grid(axis="y", linestyle="--", linewidth=0.5, alpha=0.6, color="#72a0c1")
add_stats_lines(axs[0], scores_low_sd)

# High Spread
sns.histplot(scores_high_sd, bins=4, kde=True, ax=axs[1], color="#f67280", element="bars", shrink=0.93, edgecolor='white')
axs[1].set_title("High Spread", fontsize=13)
axs[1].tick_params(axis='both', labelsize=10, colors="#54626f")
axs[1].grid(axis="y", linestyle="--", linewidth=0.5, alpha=0.6, color="#72a0c1")
add_stats_lines(axs[1], scores_high_sd)

# Minimalist styling
for ax in axs:
    for spine in ax.spines.values():
        spine.set_visible(False)

plt.suptitle("Spread Comparison: Standard Deviation", fontsize=14)
plt.tight_layout()
plt.show()
```
